2c. Clean all data

```python
# Cleaning the dataframe

ocado_df = pd.read_csv('CSV files/Ocado.csv')
ocado_df = db_clean.ocado_morrisons_cleaning(ocado_df)
ocado_master_df = pd.merge(df, ocado_df, on='generic_product')
ocado_master_df['generic_product'] = ocado_master_df['generic_product'].str.replace('%20',' ')
ocado_master_df.head(50)
```

```python
from fuzzywuzzy import fuzz, process
import spacy

def calculate_similarity(row):
    return fuzz.ratio(row['generic_product'], row['product_names'])
# Apply the function to each row in the DataFrame
ocado_master_df['fuzzy_similarity'] = ocado_master_df.apply(calculate_similarity, axis=1)

nlp = spacy.load("en_core_web_sm")
def calculate_similarity(text1, text2):
    doc1 = nlp(text1)
    doc2 = nlp(text2)
    return doc1.similarity(doc2)
ocado_master_df['spacy_similarity'] = ocado_master_df.apply(lambda row: calculate_similarity(row['generic_product'], row['product_names']), axis = 1)
```

```python
df = pd.read_csv('sainsbury.csv')
sainsbury_df = df.groupby('generic_product').head(3)
sainsbury_df.loc[:,'pre_nectar_price'] = sainsbury_df['pre_nectar_price'].replace('[^0-9.]','',regex=True)
sainsbury_df.drop('Unnamed: 0', axis=1, inplace=True)
sainsbury_df = sainsbury_df.reset_index(drop=True)
sainsbury_df['product_names'] = sainsbury_df['product_names'].str.upper()
sainsbury_df.head(20)
```

```python
df = pd.read_csv('ocado_data_complete.csv')
ocado_df = df.groupby('generic_product').head(3)
ocado_df.drop('Unnamed: 0', axis=1, inplace=True)
ocado_df = ocado_df.reset_index(drop=True)
for var in ['product_names', 'weights']:
    ocado_df[var] = ocado_df[var].str.upper()

ocado_df['product_names'] = ocado_df['product_names'] + ' ' + ocado_df['weights']
ocado_df.drop('weights', axis=1, inplace=True)
ocado_df.head(100)
```

```python
from fuzzywuzzy import fuzz, process
def find_best_match(row):
    product_name = row['product_names']
    matches = process.extract(product_name, sainsbury_df['product_names'], limit=2)
    best_match, score, _ = matches[0]
    return pd.Series({'Best_Match': best_match, 'Score': score})

ocado_df = ocado_df.head(20)
matches_df = ocado_df.apply(find_best_match, axis=1)
result_df = pd.concat([ocado_df, matches_df], axis=1)
result_df.head(30)
```
